### Q1. Explain GET and POST methods.

GET and POST are two common HTTP methods used for request-response communication between a client and a server.

##### GET

The GET method requests data from a specified resource. GET requests can be cached and remain in the browser history. It can be bookmarked. They should never be used for sensitive data and have length restrictions.

##### POST

The POST method sends data to a server to create/update a resource. POST requests are never cached and never remain in browser history. It cannot be bookmarked. The length of data is not restricted.

### Q2. Why is request used in Flask?

In Flask, a request object is used to access the incoming data from a client request. It may include mimetypes, referrers, IP addresses, raw data, HTTP methods, and headers.

GET/POST data can be recovered using the request object. The ```request.form``` attribute can be used to process form data.

### Q3. Why is redirect() used in Flask?

Flask uses the redirect() function to send a user to a specified URL with a status code. Upon calling this function, a response object is returned and redirection occurs.

### Q4. What are templates in Flask? Why is the render_template() function used?

Flask templates contain both fixed and dynamic content. They allow you to efficiently write easy-to-maintain HTML pages. Flask uses Jinja templating to render HTML templates.

The render_template() function renders a template file. It integrates Jinja with Flask. It allows us to pass data from the server to the template as keyword arguments.

````python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello World'

@app.route('/students', methods=['GET', 'POST'])
def student_api():
    if request.method == 'POST':
        data = request.get_json()
        name = data.get('name')
        age = data.get('age')
        email = data.get('email')
        return jsonify(f'Name: {name} - Age: {age} - Email: {email}')

if __name__=="__main__":
    app.run(host="0.0.0.0")
````
