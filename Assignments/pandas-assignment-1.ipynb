### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A DataFrame is a Table. It contains an array of individual entries, each with a certain value. There are rows and columns for each entry.

Series, on the other hand, are sequences of data values. If a DataFrame is a table then a Series is a list. Series are data frames with a single column. Using an index parameter, I can assign row labels to the Series. However, a Series does not have a column name.

```python
# Creating a DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
})

# Creating a Series
age_series = pd.Series([25, 30, 27])
```

### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

Pandas has several common functions to manipulate data in a DataFrame. Some of these functions include:

* head() and tail()

* info()

* describe()

* sort_values()

* groupby()

* drop()

* fillna()

Assume I have a DataFrame containing sales data, including price, date, and customer. I want to perform the following operations:

* Retrieve the first 5 rows of the DataFrame: `df.head()`

* Sort the DataFrame by sales amount in descending order: `df.sort_values('Sales Amount', ascending=False)`

* Group the data by customer name and calculate the total sales amount for each customer: `df.groupby('Customer Name')['Sales Amount'].sum()`

* Remove the 'Date' column from the DataFrame: `df.drop('Date', axis=1)`

* Fill missing values in the DataFrame with the mean value of the column: `df.fillna(df.mean())`

These are just a few examples of Pandas functions to manipulate data in a DataFrame. The choice of function depends on the specific data manipulation task I want to perform.

```python
data = {
    'Customer Name': ['Alice', 'Bob', 'Claire', 'David', 'Emma'],
    'Sales Amount': [500, 1000, 750, 400, 900],
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05']
}
df = pd.DataFrame(data)
df.head(5)
```

```python
sorted_df = df.sort_values('Sales Amount', ascending=False)
sorted_df
```

```python
grouped_df = df.groupby('Customer Name')['Sales Amount'].sum()
grouped_df
```

```python
df = df.drop('Date', axis=1)
df
```

```python
df['Sales Amount'] = df['Sales Amount'].fillna(df['Sales Amount'].mean())
df
```

### Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
name = pd.Series(['Alice', 'Bob', 'Claire'])
age = pd.Series([25, 30, 27])
gender = pd.Series(['Female', 'Male', 'Female'])

df = pd.DataFrame({
    'Name': name,
    'Age': age,
    'Gender': gender
})
```

```python
df 
```

In the above example, I created 3 separate series: name, age, and gender. Each Series represents a column in the DataFrame.

Then, I pass these 3 Series to pd.DataFrame(). DataFrame columns are represented by keys, and Series are represented by values.

Finally, I print the DataFrame, which displays the data from the multiple Series in tabular form.
