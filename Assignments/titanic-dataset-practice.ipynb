# Titanic DataSet
### Analyzing passenger data from the Titanic disaster.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("https://raw.github.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head()
```

```python
df['Name']
```

```python
# age less than 18

less_18 = df['Age'] < 18
```

```python
df[less_18]
```

```python
# sex = male & less_18

df[(df['Sex'] == 'male') & less_18]
```

```python
df.head()
```

```python
describe = df.describe()
```

```python
describe
```

```python
# mean fare
mean_fare = describe.loc['mean', 'Fare']
```

```python
# fare less than Mean fare

df[df['Fare'] < mean_fare]
```

```python
describe
```

```python
# total survived

survived = df['Survived']
```

```python
# total survived

total_survived = (survived == 1)
```

```python
df[total_survived]
```

```python
# total passengers under 18 who survived

under_18_survived = less_18 & total_survived
```

```python
df[under_18_survived]
```

```python
# total female passengers under 18 who survived

under_18_female_survived = (df['Sex'] == 'female') & under_18_survived
```

```python
# number female passengers under 18 who survived

len(df[under_18_female_survived])
```

```python
df[under_18_female_survived]
```

```python
# total male passengers under 18 who survived

under_18_male_survived = (df['Sex'] == 'male') & under_18_survived
```

```python
# number male passengers under 18 who survived

len(df[under_18_male_survived])
```

```python
df[under_18_male_survived]
```
